# librerias

```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ruta_carpeta = "/home/jovyan/work/spark/Taller Retrasos"
```

```python
lista_df = []


for archivo in os.listdir(ruta_carpeta):
    
    if archivo.endswith('.csv'):
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        
        df = pd.read_csv(ruta_archivo)
        lista_df.append(df)

df_total = pd.concat(lista_df, ignore_index=True)

```

```python
#df_total.describe()
```

```python
df_sin_nulos = df_total.dropna(axis=1, how='all')

#df_sin_nulos.describe()
```

```python
columnas_con_id = [col for col in df_sin_nulos.columns if 'id' in col.lower()]

df_sin_id = df_sin_nulos.drop(columns=columnas_con_id)

#df_sin_id.describe()
```

```python
df_sin_divs = df_sin_id[
    [
        'Year', 'DayofMonth', 'DayOfWeek', 'FlightDate',
        'Origin', 'OriginCityName', 'OriginState', 'OriginStateName',
        'Dest', 'DestCityName', 'DestState', 'DestStateName',
        'CRSDepTime', 'DepTime', 'DepDelay', 'DepDelayMinutes',
        'DepDel15', 'DepartureDelayGroups', 'DepTimeBlk', 'TaxiOut',
        'WheelsOff', 'WheelsOn', 'TaxiIn', 'CRSArrTime', 'ArrTime', 'ArrDelay',
        'ArrDelayMinutes', 'ArrDel15', 'ArrivalDelayGroups', 'ArrTimeBlk',
        'Cancelled', 'CancellationCode', 'Diverted', 'CRSElapsedTime',
        'ActualElapsedTime', 'AirTime', 'Flights', 'Distance', 'DistanceGroup',
        'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',
        'LateAircraftDelay', 'FirstDepTime', 'TotalAddGTime', 'LongestAddGTime'
    ]
]

df_sin_divs.describe()
```

```python

df_sin_divs.to_csv(ruta_carpeta + '/archivo.csv', index=False)
print("Archivo CSV guardado correctamente.")
```

# post limpieza

```python
ruta_archivo = os.path.join(ruta_carpeta, "archivo.csv")
df = pd.read_csv(ruta_archivo)
```

```python
df.columns
df_renamed = df.rename(
    columns=
    {
        "DepDelay" : "y_DepDelay", 
        "ArrDelay" : "y_ArrDelay",
        "ArrDel15" : "y_ArrDel15"
    }
)
```

```python
def diff(a, b): 
    return b - a 

def escalado(a):
    escalado = 5000/20
    cociente = int(a/escalado)
    return (cociente+1)*escalado

df_renamed['y_ArrDelay-y_DepDelay'] = df_renamed.apply( 
    lambda x: diff(x['y_DepDelay'], x['y_ArrDelay']), axis=1) 

df_renamed['DistanceGruped'] = df_renamed.apply( 
    lambda x: escalado(x['Distance']), axis=1) 
```

```python
df_renamed.columns
```

```python
columnas_y = [col for col in df_renamed.columns if col.startswith('y_')]
columnas_y.remove("y_ArrDel15")

plt.figure(figsize=(20,5))

for i in range(len(columnas_y)):
    df_grouped = df_renamed.groupby('DistanceGruped')[columnas_y[i]].mean().reset_index()
    # Crear gráfico de barras]
    plt.plot(df_grouped['DistanceGruped'], df_grouped[columnas_y[i]])
plt.legend(columnas_y)
plt.show()
```
